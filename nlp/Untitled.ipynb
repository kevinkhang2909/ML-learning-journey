```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
path = Path.home() / 'OneDrive - Seagroup/ai/kaggle_dataset/mercari-price-suggestion-challenge'
train = pd.read_csv(path / 'train.tsv', sep='\t')

print(train.shape)
train.head()
```

```python
for i, v in enumerate(['l1', 'l2', 'l3']):
    train[v] = train['category_name'].str.split('/').str[i]
    
for i in ['name', 'brand_name', 'item_description', 'l1', 'l2', 'l3']:
    train[i] = train[i].str.strip().str.lower()
    train[i].fillna('missing', inplace=True)
    
train['item_description'].replace('no description yet', 'missing', inplace=True)
train['price_log'] = np.log1p(train['price'])
```

```python
train.head()
```

```python
train.to_feather(path / 'train_clean.ftr')
```
